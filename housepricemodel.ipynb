```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

```python
home_data = pd.read_csv("Data_set.csv")
# Create target object and call it y
y = home_data.SalePrice
# Create X
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]

# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

i am  random forest regressor as my model

```python
#defing the model
model = RandomForestRegressor(random_state = 1)
```

```python
model.fit(train_X,train_y)
```

```python
y_pred = model.predict(val_X)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(val_y.values, label="Actual Price", marker='o')
plt.plot(y_pred, label="Predicted Price", marker='x')
plt.title("Actual vs Predicted House Prices")
plt.xlabel("Sample Index")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_compare = pd.DataFrame({
    "Actual": val_y.values,
    "Predicted": y_pred
})
print(df_compare.head(10))
```

Mean Absolute Error

```python
rf_val_mae = mean_absolute_error(y_pred, val_y)


print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```
